```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q1-Q3
Outliers=(df<(Q1-1.5*IQR))|((df>(Q3+1.5*IQR)))
Outliers.sum()
```

```python
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df_no_outliers
```

#### 2.Statistical Summary:

```python
mode_values=df.mode().iloc[0]
mode_values
```

```python
std_values=df.std()
std_values
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
lower_IQR=(Q1 - 1.5 * IQR)
upper_IQR=(Q3 +1.5 * IQR)
```

##### 3.Data Visualization:

```python
numerical_cols = df.select_dtypes(include=['float64']).columns
for col in numerical_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
df['NSP'] = df['NSP'].astype(float)       # convert 1.0 → 1
df['NSP'] = df['NSP'].astype('category')
df['NSP']
```

```python
df['LB_bins'] = pd.cut(df['LB'], bins=5)

df['LB_bins'].value_counts().plot(kind='bar')
plt.title("LB Binned Frequency")
plt.show()
```

```python
plt.scatter(df['LB'], df['ASTV'])
plt.title("Scatter Plot: LB vs ASTV")
plt.xlabel("LB")
plt.ylabel("ASTV")
plt.show()
```

```python
plt.scatter(df['LB'], df['AC'])
plt.title("Scatter Plot: LB vs AC")
plt.xlabel("LB")
plt.ylabel("AC")
plt.show()
```
