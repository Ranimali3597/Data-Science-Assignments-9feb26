 What happens?
TypeError: 'tuple' object does not support item assignment.
Tuples in python are immutable.
Immutable means:once a tuple is created, its values cannot be changed, added, or removed.
That's why when you try to modify students[1], python stops you and  show a type error.
In simple list= change allowed, tuple=change not allowed

###### 7.Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
x = int(input("Enter a number: "))

if x > 0:
    if x % 2 == 0:
        print("The number is positive and even.")
    else:
        print("The number is positive but odd.")
elif x == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```
